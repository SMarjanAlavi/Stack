# In which city was Python used more frequently during May to July 2018?
## (Based on StackOverflow posts in the last three months.)

```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import calendar
import datetime

# Read the data from the result of the query file
# Downloaded from "http://data.stackexchange.com/stackoverflow/query/edit/878134"
# at 2:50 pm on 2 Aug 2018.

df=pd.read_csv("python2018Q2.csv", sep=',')

#print (df.head())
locations = set(df['Location'].tolist())
#print(locations)
df_Date_Location_TagName = df.loc[:,['Date','Location', 'TagName']]
print('We are investigating', len(df_Date_Location_TagName), 'posts between', min(df_Date_Location_TagName['Date']),'to', max(df_Date_Location_TagName['Date']),\
      'from', len(locations), 'locations.\n')
```

# In which city was Python used more frequently during May to July 2018?

```python
#print (df_Date_Location_TagName)
df_Location_TagName = df_Date_Location_TagName.drop(columns = 'Date')
df_Location_group = df_Location_TagName.groupby(['Location']).agg({'count'}).sort_values(('TagName', 'count'), ascending=False)
#print(df_Location_group)
# Top 20 loctions where python tags were last used.
df_top_locations = df_Location_group.head(20)
print(df_top_locations)
Top_Locations = df_top_locations.index.tolist()
#print(Top_Locations)
```

```python
print('Top locations: ', Top_Locations)

print('\nAs it can be seen from the chart,', df_top_locations[('TagName','count')].idxmax(), 'is the top location most for Python.\n')

fig = df_top_locations.plot.bar(title = 'Top python cities betweenin May to July 2018.');
fig.legend_.remove();
fig.set_xlabel("Location");
fig.set_ylabel("Number of python Tags\n in the StackOverflow posts");
```
